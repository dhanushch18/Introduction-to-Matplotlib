```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,3]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.plot(x,y,color="green",linestyle="dashdot")
```

```python
plt.plot(x,y,'--r*')
```

```python
plt.plot(x,y,'--r*',alpha=0.6) #alpha for opacity
```

```python
days=[1,2,3,4,4,5,6,77]
max=[4,8,8,77,88,89,99,75]
min=[3,5,6,55,64,23,55,66]
avg=[3,6,66,55,54,33,77,99]
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max,label="Max")
plt.plot(days,min,label="Min")
plt.plot(days,avg,label="Avg")

plt.legend(loc="best")
plt.grid()
```
