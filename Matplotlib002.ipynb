```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['FB','WP','IG']
revenue=[90,77,120]
profit=[60,50,90]
```

```python
x=np.arange(len(company))
```

```python
x
```

```python
plt.bar(x,revenue)
```

```python
plt.xticks(x,company)
plt.ylabel('revenue')
plt.bar(x,revenue,label='revenue')
plt.bar(x,profit,label='profit')
plt.legend()
```

```python
plt.xticks(x,company)
plt.ylabel('revenue')
plt.bar(x-0.2,revenue,width=0.4,label='revenue')
plt.bar(x+0.2,profit,width=0.4,label='profit')
plt.legend()
```

```python
plt.yticks(x,company)
plt.ylabel('revenue')
plt.bar(x-0.2,revenue,label='revenue')
plt.bar(x+0.2,profit,label='profit')
plt.legend()
```
